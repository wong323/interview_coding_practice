```python
import unittest
```

```python
class solution(object):
    
    def compute_area(self,x, y, n):
        if x > y:
            area = y * n
        else:
            area = x * n
        return area
    
    def max_area(self, l):
        maxarea = 0
        
        for j in range(0, len(l)):
            
            end = len(l) - j
            
            for i in range(j, end):
                n = i - j + 1
                area = self.compute_area(l[j], l[j+i], n)
                
                if area > maxarea:
                    maxarea = area  
        
        return maxarea
```

```python
class solutiontestcase(unittest.TestCase):
    def setUp(self):
        self.test1 = solution()
        
    def test_maxarea(self):
        l = [1,8,6,2,5,4,8,3,7]
        answer = 49
        self.assertEqual(self.test1.max_area(l), answer, msg='this is wrong')
        
unittest.main(argv=[''], verbosity=2, exit=False)
```
