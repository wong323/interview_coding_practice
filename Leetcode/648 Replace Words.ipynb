```python
import unittest
```

```python
class test_solutino(unittest.TestCase):
    
    def setUp(self):
        self.s = solution()
        
    def test_solution(self):
        d = ["cat", "bat", "rat"]
        sentence = "the cattle was rattled by the battery"
        answer = "the cat was rat by the bat"
        self.assertEqual(self.s.replace_root(d, sentence), answer, msg='this is wrong')
    
    def test_solution_empty(self):
        d = []
        sentence = "the cattle was rattled by the battery"
        answer = "the cattle was rattled by the battery"
        self.assertEqual(self.s.replace_root(d, sentence), answer, msg='this is wrong')
     
    def test_solution_multimatch(self):
        d = ["cat", "catt", "rat"]
        sentence = "the cattle catt was rattled by the battery"
        answer = "the cat cat was rat by the battery"
        self.assertEqual(self.s.replace_root(d, sentence), answer, msg='this is wrong')
        
        
    def test_solution_hard(self):
        d = ["a", "b", "c"]
        sentence = "aadsfasf absbs bbab cadsfafs"
        answer = "a a b c"
        self.assertEqual(self.s.replace_root(d, sentence), answer, msg='this is wrong')    
        
unittest.main(argv=[''], verbosity=2, exit=False)    
```

```python
class solution(object):
    
    def replace_root(self, d, sentence):
        list_words = sentence.split()
        new_list_words = []
        
        for word in list_words:
            boolean = True
            for root in d:
                if root in word[0:len(root)]:
                    new_list_words.append(root)
                    boolean = False
                    break
                else:
                    pass
                
            if boolean == True:
                
                new_list_words.append(word)
                
        output = " ".join(new_list_words)
        return output
```
