```python
class Solution:
    
    def __init__(self, grid):
        self.grid = grid
        self.row = len(grid)
        self.col = len(grid[0])
        
    def numIslands(self):
        """
        :type grid: List[List[str]]
        :rtype: int
        """
        
        row = self.row
        col = self.col
        
        total_number_island = 0
        for i in range(row):
            for j in range(col):
                
                value = self.grid[i][j]
                if value == 1:
                    total_number_island += 1
                    self.set_adj_zero(i , j)
        return total_number_island
    
    def set_adj_zero(self, i, j):        
        if i < 0 or j < 0 or i > self.row-1 or j >self.col-1:
            return
        
        if self.grid[i][j] == 0:
            return
        
        if self.grid[i][j] == 1:
            self.grid[i][j] = 0
            self.set_adj_zero(i, j-1)
            self.set_adj_zero(i, j+1)
            self.set_adj_zero(i-1, j)
            self.set_adj_zero(i+1, j)
```
