```python
import unittest
```

```python
class test_solution(unittest.TestCase):
    
    def setUp(self):
        self.s = solution()
        
    def test_remove_k_largest(self):
        ar = [3,2,1,5,6,4]
        ans = 5
        k = 2
        self.assertEqual(self.s.find_k_largest(ar, k), ans, msg='this is wrong')
        
    def test_remove_k_largest_dup(self):
        ar = [3,2,3,1,2,4,5,5,6]
        ans = 4
        k = 4
        self.assertEqual(self.s.find_k_largest(ar, k), ans, msg='this is wrong')

unittest.main(argv=[''], verbosity=2, exit=False)        
```

```python
class solution(object):
    
    def find_k_largest(self, ar, k):
        ar.sort()
        rev_ar = ar
        rev_ar.reverse()
        
        answer = rev_ar[k-1]
        print(rev_ar)
        return answer
        
```
