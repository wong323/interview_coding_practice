```python
import unittest
from math import inf
```

```python
class test_solution(unittest.TestCase):
    
    def setUp(self):
        self.s = Solution()
        
    def test_rob(self):
        tc =  [-2,-3,4,-1,-2,1,5,-3]
        ans = 7
        self.assertEqual(self.s.find_max_sum(tc), ans , msg='this is wrong')

unittest.main(argv=[''], verbosity=2, exit=False)   
```

```python
class Solution(object):
    
    def __init__(self):
        pass
    
    def find_max_sum(self, l):
        
        max_so_far = -inf
        max_ending_here = 0

        for i in range(0, len(l)): 
            max_ending_here = max_ending_here + l[i] 
            if (max_so_far < max_ending_here): 
                max_so_far = max_ending_here 

            if max_ending_here < 0: 
                max_ending_here = 0   
        return max_so_far 
 
```
