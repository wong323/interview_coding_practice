```python
import unittest
```

```python
class Solution(object):
    
    def find_sub(self, l, tcsum):
        
        # create 2 loops to iterate every possible combo to see if match
        for i in range(0, len(l)):
            temp_sum = 0
            for j in range(i, len(l)):
                temp_sum += l[j]
                
                if temp_sum == tcsum:
                    return (i,j)
            
        return "no ans"
    
    def find_sub_efficient(self, l, tcsum):
        
        temp_sum = 0
        start_index = 0
        
        # only 1 loop, when temp_sum gets larger than tcsum, then will drop the previous values
        for i in range(0, len(l)):
            temp_sum += l[i]
            
            while temp_sum > tcsum:
                temp_sum -= l[start_index]
                start_index += 1
                
            if temp_sum == tcsum:
                return (start_index, i)
            
        return "no ans"
                
```

```python
class test_solution(unittest.TestCase):
    def setUp(self):
        self.s = Solution()
        
    def test_find_sub(self):
        tc = [1,4,20,3,10,5]
        tcsum = 33
        ans = (2,4)
        self.assertEqual(self.s.find_sub(tc, tcsum), ans, msg='this is wrong')
        
        
    def test_find_sub_efficient(self):
        tc = [1,4,20,3,10,5]
        tcsum = 33
        ans = (2,4)
        self.assertEqual(self.s.find_sub_efficient(tc, tcsum), ans, msg='this is wrong')
unittest.main(argv=[''], exit= False)
```
