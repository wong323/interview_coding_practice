```python
import unittest
```

```python
class Solution(object):
    def __init__(self):
        pass
    
    def find_large(self, l):
        
        # 2 for loops to test every combo, if greater than max_length then update
        max_length = 0        
        for i in range(0, len(l)):
            temp_length = 0
            counter = 0
            
            for j in range(i, len(l)):
                
                if l[j] == 1:
                    counter += 1
                else:
                    counter -= 1

                if counter == 0:
                    temp_length = j-i
                    
                    if temp_length > max_length:
                        max_length = temp_length
                        ans = (i,j)  
        
        if ans:
            return ans
        else:
            return 'no ans'
    
    def find_large_efficient(self, l):
        
        # using 1 for loop
        hash_map = {}
        curr_sum = 0  
        max_len = 0  
        ending_index = -1 
  
        # first replace all 0 with -1, so now we solving curr_sum == 0 
        for i in range (0, len(l)): 
            if(l[i] == 0): 
                l[i] = -1; 
            else: 
                l[i] = 1; 
        
        for i in range(0, len(l)):
            curr_sum += l[i]
            
            # if curr_sum is zero, update max_len and ending_index
            if curr_sum == 0:
                max_len = i + 1
                ending_index = i
                
            if curr_sum in hash_map:
                max_len = max(max_len, i - hash_map[curr_sum])
                ending_index = i
            else:
                hash_map[curr_sum] = i
                
        print(max_len)
        if max_len == 0:
            return "no ans"
        else:
            return (ending_index - max_len + 1 , ending_index)
    
```

```python
class test_solution(unittest.TestCase):
    def setUp(self):
        self.s=Solution()
        
    def test_find_large(self):
        tc = [1,0,1,1,1,0,0]
        ans = (1,6)
        self.assertEqual(self.s.find_large(tc), ans, msg = 'this is wrong')
        
    def test_find_large_efficient(self):
        tc = [1,0,1,1,1,0,0]
        ans = (1,6)
        self.assertEqual(self.s.find_large_efficient(tc), ans, msg = 'this is wrong')

    def test_find_large_efficient2(self):
        tc = [0,1,1,1,0,0]
        ans = (0,5)
        self.assertEqual(self.s.find_large_efficient(tc), ans, msg = 'this is wrong')
unittest.main(argv=[''], exit = False)
```
