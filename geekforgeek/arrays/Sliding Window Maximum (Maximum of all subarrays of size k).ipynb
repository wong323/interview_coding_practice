```python
from math import inf
import unittest
```

```python
class Solution(object):
    def __init__(self):
        pass
    
    def find_max(self, l , k):
        
        # for every possible starting point, iterate from it to k steps, and find max
        ans_list = []
        for i in range(0, len(l)-k+1):
            temp_max = -inf
            for j in range(i, i+k):
                if l[j] > temp_max:
                    temp_max = l[j]
            ans_list.append(temp_max)
                
        return ans_list
    
    def find_max_efficient(self, l, k):
        
        ## to do lol
        ans_list = []
        
        temp_max = -inf
        for i in range(0, k):
            if l[j] > temp_max:
                    temp_max = l[j]
            ans_list.append(temp_max)
        last_max = temp_max
            
        for i in range(k, len(l)-k+1):
            
            if l[i] > last_max:
                last_max = l[i]
            
        
        return ans_list
```

```python
class test_solution(unittest.TestCase):
    
    def setUp(self):
        self.s = Solution()
        
    def test_find_max(self):
        tc = [1,2,3,1,4,5,2,3,6]
        k = 3
        ans = [3,3,4,5,5,5,6]
        self.assertEqual(self.s.find_max(tc, k), ans, msg = 'this is wrong')
      
    def test_find_max_efficient(self):
        tc = [1,2,3,1,4,5,2,3,6]
        k = 3
        ans = [3,3,4,5,5,5,6]
        self.assertEqual(self.s.find_max_efficient(tc, k), ans, msg = 'this is wrong')
        
unittest.main(argv=[''], exit = False)
```
